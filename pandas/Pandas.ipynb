### Manipulating dataframe
***

```python
# adding a column to a dataframe
carprice_df['Price2'] = carprice.price
# here the data of price column is stored into the new column price2
```

```python
# dropping a row
carprice_df.drop(index=0, axis=0) 

# axis=0 for dropping row and axis=1 for dropping column
```

```python
#we could mention the name of column to drop
carprice_df.drop(columns = 'fueltype', axis=1)
```

```python
# locating a row using the index value

carprice_df.iloc[2]
```

```python
# displying specific column by indexing
carprice_df.iloc[:,2]
```
