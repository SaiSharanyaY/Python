## Functions
***

1. syntax : def func_name(parameters):
                   indentated code

```python
# example - 1

def minimum(first, second):
    if (first < second):
        return first
    return second

# example - 2 : fibonocci series : recursive case

def fib(n):
    # The base cases
    if n <= 1:  # First number in the sequence
        return 0
    elif n == 2:  # Second number in the sequence
        return 1
    else:
        # Recursive call
        return fib(n - 1) + fib(n - 2)


print(fib(6))
```

```python
# Lists 

# This is the most common used, can store diff data types, enclosed in [] and is mutable.

a=["one", 20, True]

# list - ception (list in a list)
aa=[[2000,"march"],[2009,"april"],[2010,"may"],[2020,"june"]]
print(aa[0])
print(aa[0][1])
print(aa[0][1][2])

# lists can be merged using + and can be extended using extend()
x=[1,2,3,4,5]
y=[6,7,8,9,10]
z=x+y
x.extend(y)
print(z)
print(x)

# appending a list using append()
x.append(12)

#inserting a new element at a specific index insert(index, value)
x.insert(3,14)

# removing element is pop() 
x.pop()

#removing element at specific index remove(element)
x.remove(5)

#list slicing
print(x[2:8:2])

# index search using index(value)
x.index(3)

#To verify if an element is in the list or not - membership operators : 
print(2 in x)
print(5 not in x)

# sorting the list 
print(x.sort())
```

```python
# creating a list comprehension

n=[10,20,30,40,50]
n1=[]
i=0
for i in n:
    n1.insert(i,n*2)

n2=[n*2 for i in n]
print(n)
print(n2)


list1 = [30, 50, 110, 40, 15, 75]
list2 = [10, 60, 20, 50]

sum_list = [(n1, n2) for n1 in list1 for n2 in list2 if n1 + n2 > 100]

print(sum_list)
```

```python
# Tuple

# tuples are immutable, enclosed in () , are ordered and follow linear notation.
a=("one",2,3.5,True)

#length - len()
print(len(a))

#index
print(a[2])

#slicing
print(a[2:6])

#Merging tuples
aa=("two",4,6.7,False)
aaa=a+aa
print(aaa)

#Nested tuples
aaaa=(a,aa)
print(aaaa)

#element in tuple : boolean
print("one" in a )

#index of an element
print(a.index("one"))

#reversing a tuple
print(a[::-1])

#count() method
print(a.count())

#sorted()
x=(2,3,4,7,6,5)
print(sorted(x))
print(sorted(x, reverse=True))

# due to immutability no elements can be inserted or removed or editted.
```

### Differences between Lists and Tuples
1. Lists are mutable but tuples are immutable.
2. Lists are enclosed in [] and tuples in ()
3. Tuples are faster than lists
4. Tuples rae preffered for heterogenoues collections while lists are preferred for homogenous collections
5. Dictionaries can use any immutable object as key. So, tuples can be used as dictionary keys when needed but lists cannot be used.

```python
# Dictionary

#A dictionary stores key-value pairs, where each unique key is an index which holds the value associated with it.
#They are unordered, non linear structure and are enclosed in {} brackets.

# dict() constructor can be used to construct a dictionary.
# keys will be automatically converted to strings.
#all keys should be unique.
phone=dict(raj=1109,rani=9877,karem=7654,quil=3456)
print(type(phone))
 
#Accessing values
print(phone.get("raj"))

# Adding and Updating entries :
phone["kuen"]=7289
phone["raj"]=8888

#removing entry
del phone["raj"]
# or
phone.pop("raj")
# or 
phone.popitem() # this is used to pop items randomly in older versions than 3.7

# length
print(len(phone))

#checlking key existence
print("rani" in phone)

#updating i.e addn of new dict
b=[one=1, two=2]
phone.update(b)
print(phone)

# Dictionary Comprehension
# However, to iterate the dictionary, we’ll use the dict.items() 
# operation which turns a dictionary into a list of (key, value) tuples.

houses = {1: "Gryffindor", 2: "Slytherin", 3: "Hufflepuff", 4: "Ravenclaw"}
new_houses = {n**2: house + "!" for (n, house) in houses.items()}
print(houses)
print(new_houses)
```

```python
# Sets

#set is an unordered collection of elements, so indexing cant be used.
# mutable data structures cannot be added to a set, so it's fine adding a tuple to it.
# It's used to keep perfect track of elements and to avoid duplicacy.
# they are enclosed in {}

a=set({"0n2",23,("one","two")})
# or
b = {23, 76, "three"}
print(len(a))
print(type(b))

# Element addition can be done using add() or update() but update is used only for list,tuple,set,strings.
e=set()
e.add(65)
e.update(b)
print(e)

# remove() and discard() are used to delete elements but here remove() creates error when item is not found unlike discard()
e.remove(65)
e.discard(65)

# set iteration
odd_list = [1, 3, 5, 7]
unordered_set = {9, 10, 11, 12, 13, 14, 15, 16, 17}

print(unordered_set)

for num in unordered_set:
    if(not num % 2 == 0):
        odd_list.append(num)

print(odd_list)


# Set operations

#Union - can be performed using | or union() method.

set_A = {1, 2, 3, 4}
set_B = {'a', 'b', 'c', 'd'}

print(set_A | set_B)
print(set_A.union(set_B))
print(set_B.union(set_A))

# Intersection - can be performed using & or intersection() method.

set_A = {1, 2, 3, 4}
set_B = {2, 8, 4, 16}

print(set_A & set_B)
print(set_A.intersection(set_B))
print(set_B.intersection(set_A))

# Difference - collecn of unique elements present in a and not in b, - or difference() is used.

set_A = {1, 2, 3, 4}
set_B = {2, 8, 4, 16}


print(set_A - set_B)
print(set_A.difference(set_B))

print(set_B - set_A)
print(set_B.difference(set_A))
```
